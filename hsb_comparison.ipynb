```python
import pickle 

def load_pkl(file):
    with open(file, 'rb') as file:
        return pickle.load(file)
    
```

```python
ROOT_DIR = '/home/leostre/Рабочий стол/py-boost/sketchboost-paper/runs/SKETCHBOOST_ALLK'

DATASETS = ['helena', 'moa', 'rf1']

K = [1, 10, 20]

MODELS = ['hb', 'pb', 'hb15']

FOLDS = list(range(5))
```

```python
from itertools import product
import os 
import pandas as pd

all_results = []
KEYS = ['dataset', 'k', 'model']

for dataset, k, model, fold in product(DATASETS, K, MODELS, FOLDS):
    path = os.path.join(ROOT_DIR, dataset, model, f'topk_{k}', f'fold_{fold}', 'results.pkl')
    if not os.path.exists(path):
        print(f'Skipped {path}')
        continue
    
    results = load_pkl(path)
    all_results.append(
        dict(zip(KEYS, (dataset, k, model))) | results
    )

all_results = pd.DataFrame(all_results) 
```

```python
def get_one_field(all_data, field, value, aggfunc=None):
    one_model = all_data[all_data[field] == value]
    index = [c for c in KEYS if c != field]
    one_model.drop(field, axis=1, inplace=True)
    if aggfunc:
        one_model = one_model.groupby(index).aggregate(aggfunc)

    return one_model
```

```python
import numpy as np 

def agg_mean_std(x):
    return f'{np.mean(x):.2f}±{np.std(x):.2f}'
```
